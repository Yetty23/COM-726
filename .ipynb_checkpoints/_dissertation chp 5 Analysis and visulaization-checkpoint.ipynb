```python
import pandas as pd

# Load the dataset
file_path = 'heart_failure_clinical_records_dataset.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the style for seaborn
sns.set(style="whitegrid")

# Plot 1: Age Distribution of Patients
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=20, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plot 2: Serum Creatinine Levels by Death Event
plt.figure(figsize=(10, 6))
sns.boxplot(x='DEATH_EVENT', y='serum_creatinine', data=data)
plt.title('Serum Creatinine Levels by Death Event')
plt.xlabel('Death Event')
plt.ylabel('Serum Creatinine (mg/dL)')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# Plot 3: Correlation Matrix
plt.figure(figsize=(12, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of the Heart Failure Clinical Dataset')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from lifelines import KaplanMeierFitter

# Load the dataset
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')

# 1. Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# 2. Distribution of Features
df.hist(figsize=(15, 12), bins=15)
plt.suptitle('Distribution of Features')
plt.show()

# 3. Boxplot for Numerical Features
plt.figure(figsize=(15, 8))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title('Boxplot for Numerical Features')
plt.show()

# 4. Survival Analysis (Kaplan-Meier Curve)
kmf = KaplanMeierFitter()

T = df['time']  # Duration of time
E = df['DEATH_EVENT']  # Event occurred or censored

kmf.fit(T, event_observed=E)

plt.figure(figsize=(10, 6))
kmf.plot()
plt.title('Kaplan-Meier Survival Curve')
plt.xlabel('Time (days)')
plt.ylabel('Survival Probability')
plt.show()

# 5. Bar Plot for Categorical Features vs Death Event
categorical_features = ['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking']

for feature in categorical_features:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=feature, hue='DEATH_EVENT', data=df, palette='Set1')
    plt.title(f'{feature.capitalize()} vs DEATH_EVENT')
    plt.show()

# 6. Pair Plot for Selected Features
sns.pairplot(df[['age', 'ejection_fraction', 'serum_creatinine', 'platelets', 'creatinine_phosphokinase', 'DEATH_EVENT']], hue='DEATH_EVENT', palette='Set1')
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()

# 7. Violin Plot for Continuous Features
continuous_features = ['age', 'creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium']

for feature in continuous_features:
    plt.figure(figsize=(8, 5))
    sns.violinplot(x='DEATH_EVENT', y=feature, data=df, palette='Set1')
    plt.title(f'{feature.capitalize()} Distribution by Death Event')
    plt.show()
```
